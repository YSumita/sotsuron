```python
def visualize_resampled_data(data, label = None, figsize=(10, 10)):
    '''
    90 * 70のデータを可視化する
    '''
    
    plt.figure(figsize=(10,10))

    # 陸地の描画
    m = Basemap(projection='npstere', boundinglat=49.9, lon_0=90, resolution='l')
    m.drawcoastlines()
    m.fillcontinents()
    # 見栄えのために海の色をcmapの色と合わせた
    m.drawmapboundary(fill_color='#143067')

    # 密接度の描画
    # おおよそこんな感じだと思う(調整した)
    x = np.linspace(3300000, 7600000, data.shape[1])
    y = np.linspace(2100000, 7550000, data.shape[0])
    xx, yy = np.meshgrid(x, y)

    cmap = plt.get_cmap("Blues")
    m.pcolormesh(xx, yy, data, cmap=cmap)
    m.colorbar(location="bottom", label=label)

    plt.show()
```
